## Import library

## Load simulation directory

```python
data = emout.Emout('simulation directory')
```

```python
data.inp
```

```python
data.inp.jobcon.nstep, data.inp.nstep
```

```python
print(data.j1y[-1].shape)
```

## 2d plot

```python
data.j1y[-1, 0:308, int(data.inp.ny//2), :].plot()
```

```python
data.nd2p[-1, :, int(data.inp.ny//2), :].plot(mode='cmap')
```

## Contour plot

```python
data.phisp[-1, :, int(data.inp.ny//2), :].plot(mode='cont')
```

## Colormap + Contour plot

```python
data.nd1p[-1, :, :, int(data.inp.nx//2)].plot(mode='cmap+cont')
```

## 1D plot

```python
data.phisp[-1, :, int(data.inp.ny//2), int(data.inp.nx//2)].plot()
```

## Animation

```python
data.phisp[:, :, int(data.inp.ny//2), :].gifplot(to_html=True)
```

## Streamline plot

```python
data.jyz[-1, :, :, int(data.inp.nx//2)].plot(mode='stream', cmap='jet')
```

```python
data.phisp[-1, :, :, int(data.inp.nx//2)].plot(mode='cmap')
data.jyz[-1, :, :, int(data.inp.nx//2)].plot(mode='stream')
```
